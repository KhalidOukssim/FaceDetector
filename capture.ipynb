### you can remove print it is just For Test

```python
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
video=cv2.VideoCapture(0)
while True:
        check, frame = video.read()
        #print(check) #if exist
        #print(frame) #the array
        print("pass0")


        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        print("pass1")
        faces=face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5)
        print("pass2")
        #start from the originale size and decrese the image for 5% and search for bigger faces
        for x,y,w,h in faces:
            gray=cv2.rectangle(gray,(x,y),(x+w,y+h),(0,255,0),3)
        #time.sleep(3)
        print("pass3")
        cv2.imshow("Capturing",gray)
        print("pass4")
        key=cv2.waitKey(1)
        if key==ord('q'):
            break
        
video.release()
cv2.destroyAllWindows
```
